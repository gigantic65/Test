```python
import pandas as pd
```

```python
import re

f = open('Incan_2.csv', 'r', encoding = 'ISO-8859-1')
states_list = []
result = []
name = []
pro = []
add = []
add_v = []


for line in f.readlines():
    line.strip()
    states_list.append(line.strip()) # strips newline character at the end of the line
    
for i in states_list:
    i = i.replace('|', '  ')
    print(i)
    check_name = re.search(r'([\@CLR]+)(.+)', i)
    print(check_name.group(1))
    check_add = re.search(r'([\@CNT]+)(.+)', i)
    print(check_add.group(1))
    #i = re.sub('[^a-zA-Z]', ' ', i)
    if check_name.group(1)=='@CLR':
        #i = re.sub('[./()]', '', i)
        pro.append(re.findall('[" "][a-zA-Z_][-][a-zA-Z0-9_]+[" "]', i))
        
        name.append(re.findall('[A-Za-z0-9-+][-].+[" "]', i))
        
        
        #elif re.findall('[A-Za-z0-9-+][-][A-Za-z0-9-+][-].+[-][A-Za-z0-9-+]+', i) ! = None:
        #name.append(re.findall('[A-Za-z0-9-+][-][A-Za-z0-9-+][-].+[-][A-Za-z0-9-+]+', i))
            
        #if re.findall('[A-Za-z0-9-+][-][A-Za-z0-9-+]+[-][A-Za-z0-9-+][-].+[-][A-Za-z0-9-+]+', i) !=None:
        #    name.append(re.findall('[A-Za-z0-9-+][-][A-Za-z0-9-+]+[-][A-Za-z0-9-+][-].+[-][A-Za-z0-9-+]+', i))    

          
        print(pro)
        print(name)
        
    if check_add.group(1)=='@CNT':
        add.append(re.findall('[S][a-zA-Z0-9_]+', i))
        add_v.append(re.findall('[0-9_]+[.][0-9_]+', i))
        print(add)
        print(add_v)
```

```python
name = pd.DataFrame(name,columns=['check'])
name_split = name["check"].str.split(" ")
name["Name"] = name_split.str.get(0)
name["Product"] = name_split.str.get(2)
name
```

```python
list = ['SK1', 'SK2', 'SB1', 'SB2', 'SG1', 'SY1', 'SY2', 'SY3', 'SO1', 'SP1', 'SV1', 'SR1', 'SR2', 'SR3', 'SW1']
test = name.drop(['check'], axis=1)
for column in list:
    test[column] = 0.0
test
```

```python
add = pd.DataFrame(add)
add_v = pd.DataFrame(add_v)

for i in range(add.shape[0]):
    for j in range(add.shape[1]):
        if add.iloc[i][j] in test.columns:
            test.loc[i,add.iloc[i][j]]=add_v[j][i]
print(test)
```
